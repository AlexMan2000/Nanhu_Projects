```python
import pandas as pd
```

```python
data = pd.read_csv("./项目总评.csv")
```

```python
data["左边关节最大冲击指数"] = data["左边关节最大冲击指数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["右边关节最大冲击指数"] = data["右边关节最大冲击指数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["左边关节平均冲击指数"] = data["左边关节平均冲击指数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["右边关节平均冲击指数"] = data["右边关节平均冲击指数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["左边关节损伤风险系数"] = data["左边关节损伤风险系数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["右边关节损伤风险系数"] = data["右边关节损伤风险系数"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["速度变化(m/s)"] = data["速度变化(m/s)"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["加速度变化(m/s2)"] = data["加速度变化(m/s2)"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["重心向右变化"] = data["重心向右变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["左上肢稳定性变化"] = data["左上肢稳定性变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["右上肢稳定性变化"] = data["右上肢稳定性变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["左下肢稳定性变化"] = data["左下肢稳定性变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["右下肢稳定性变化"] = data["右下肢稳定性变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
data["重心向前变化"] = data["重心向前变化"].str.split(",").apply(lambda x:list(map(lambda y:float(y), x)))
```
